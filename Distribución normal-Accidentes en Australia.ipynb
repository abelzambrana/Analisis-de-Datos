# Realizaremos la Distribución normal de Crash_Year

```python
from scipy import stats
import numpy as np
```

Calculamos la media

```python
mu = df1['Crash_Hour'].mean()
mu
```

Calculamos la desviación standar

```python
sigma = df1['Crash_Hour'].std()
sigma
```

```python
valores= df1['Crash_Hour']
m_valores= np.array(valores)
m_valores
```

Calculamos los valores Z

```python
z = (m_valores - mu)/sigma
z
```

```python
df_z = pd.DataFrame(z)
df_z
```

Acontinuación obtenemos la grafica de la variable Crash_Hour, aunque lamentamente no sigue una distribución normal.

```python
#datos = np.random.normal(mu,sigma,10000)    #histograma de datos simulados
plt.figure(figsize=(25,10))
plt.hist(z,50, color="r", density = True, alpha=0.7)  #le doy la continuidad con la cantidad de intervalos
plt.xlabel('valores')
plt.ylabel('frecuencia')
plt.title('Histograma Normal')
plt.show()
```

# Prueba de Independencia, entre Crash_Day_Of_Week y Crash_Severity en el año 2008

```python
año_2008 = df1 ['Crash_Year'].isin([2008])
df_2008 = df1[año_2008]
df_2008
```

```python
df_Crash_Day_Of_Week = df_2008['Crash_Day_Of_Week']
df_Crash_Severity = df_2008['Crash_Severity']
```

```python
df_nuevo = pd.concat([df_Crash_Day_Of_Week , df_Crash_Severity], axis=1)
df_nuevo
```

```python
contigency= pd.crosstab(df_nuevo['Crash_Day_Of_Week'], df_nuevo['Crash_Severity'])
contigency
```

```python
# Porcentajes por filas:
contigency_pct = pd.crosstab(df_nuevo['Crash_Day_Of_Week'], df_nuevo['Crash_Severity'], normalize='index')
contigency_pct
```

En el siguiente grafico podemos ver la fuerte dependencia entre la categoria Properly damage only de la variable Crash_Severity con la categoria Friday de la variable Crash_Day_of_week.

```python
plt.figure(figsize=(12,8))
sns.heatmap(contigency, annot=True, cmap="YlGnBu")
```
